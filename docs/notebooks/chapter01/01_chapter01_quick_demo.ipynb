토큰화, 규칙 기반 분류, SQLite 키워드 검색을 빠르게 확인합니다.


## 1. 간단한 토큰화

```python
sentence = "나는 자연어처리를 공부하고 있습니다."

tokens = sentence.replace(".", "").split()

print("원문:", sentence)
print("토큰:", tokens)
```

## 2. 규칙 기반 의도 분류

```python
def detect_intent(text):
    if "환불" in text or "반품" in text or "취소" in text:
        return "환불 문의"
    if "배송" in text or "택배" in text or "도착" in text:
        return "배송 문의"
    if "결제" in text or "카드" in text or "입금" in text:
        return "결제 문의"
    if "비밀번호" in text or "로그인" in text:
        return "계정 문의"
    return "기타 문의"

sentences = [
    "상품 환불은 어떻게 하나요?",
    "배송이 아직 안 왔어요.",
    "카드 결제가 실패했습니다.",
    "비밀번호를 잊어버렸어요.",
    "상담원과 통화하고 싶어요.",
]

for sentence in sentences:
    print(sentence, "→", detect_intent(sentence))
```

## 3. SQLite 키워드 검색

```python
import sqlite3

conn = sqlite3.connect(":memory:")
cur = conn.cursor()

cur.execute("""
CREATE TABLE documents (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    content TEXT NOT NULL
)
""")

documents = [
    ("휴가 규정", "연차휴가는 입사일 기준으로 산정합니다."),
    ("배송 안내", "상품은 결제 완료 후 2일 이내 출고됩니다."),
    ("환불 안내", "환불은 상품 수령 후 7일 이내에 신청할 수 있습니다."),
]

cur.executemany("INSERT INTO documents (title, content) VALUES (?, ?)", documents)
conn.commit()

keyword = "휴가"
cur.execute("SELECT title, content FROM documents WHERE content LIKE ?", (f"%{keyword}%",))
rows = cur.fetchall()
rows
```

SQLite의 `LIKE` 검색은 특정 단어가 포함된 문서를 찾는 데 유용합니다. 하지만 의미가 비슷한 표현까지 찾으려면 임베딩과 벡터 검색이 필요합니다.
